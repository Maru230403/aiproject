# 로컬 저장소 - git

## git init

- vs code 로 프로젝트 root 폴더를 open
  - Source Control TAB 에서 Initialize Repository 클릭
    - .git 숨김 폴더 생성됨
    - CLI : git init 과 같은 기능

### .git

- 프로젝트 root 폴더에 생성되는 숨김폴더
  - git 이 버전 관릴를 하기 위한 메타 정보들
  - 이 디렉토리를 지우면 모든 변경 이력은 없어짐

## .gitignore

- git 의 버전관리에서 제외하는 항목 리스트 파일
  - 최초 commit 전 구성해 놓아야 환경설정 파일등 불필요한 파일들이 관리되는 것을 막을 수 있음
  
```
# 개발환경 폴더
**/env/
**/node_modules/

# 개발환경 파일
.env
```

image.png

```
# 확장자가 .js 인 파일은 무시
*.js

# .js 파일들은 모두 무시되지만, test.js만은 무시하지 않음
!test.js

# 현재 디렉토리에 있는 /test.js파일은 무시되지만,
# subDir/test.js 같이 특정 디렉토리 하위에 있는 test.js는 무시되지 않음
/test.js

# node_modules/ 디렉토리에 있는 모든 파일을 무시
node_modules/

# src/ 하위의 .js파일만 무시
src/*.js

# src/ 하위에 존재하는 모든 디렉토리의 .txt 파일을 무시
src/**/*.txt

# 현재 디렉토리와 그 하위 디렉토리 내에 존재하는 모든 .js 파일을 무시
/**/*.js

# 현재 디렉토리 내에 존재하는 모든 .js .ts 파일 무시
/*.{js, ts}

# 현재 디렉토리 내에 있는 ex1.js ex2.js ex3.js 파일 무시
/ex[1-3].js
```

- add 된 파일 or commit 된 파일을 제거하고 .gitignore 파일에 등록
  - git rm <파일명>
    - 원격저장소와 로컬저장소에 있는 파일을 삭제
  - git commit -m "fixed untracked files"
  - .gitignore 파일에 삭제할파일명 등록

## git config

### 설정 범위

- local(지역)
  - 특정 저장소로 한정(특정 폴더)
  - .git\config 에 저장

- global(전역)
  - 현재 사용자(User)의 모든 저장소로 한정
  - `\Users\<user>\.gitconfig` 에 저장

- system(시스템)
  - 현재 컴퓨터의 모든 사용자의 모든 저장소에 적용
  - C:\ProgramData\Git\config 에 저장

### 속성 값 설정

- git config --<범위> <속성이름> <값>
  - <범위> - local 은 표기 생략

### 속성 값 읽기

- git config --<범위> <속성이름>

### 속성 값 삭제

- git config --<범위> --unset <속성이름>

### user.email, user.name

- 처음으로 로컬 저장소에 commit 하기 전 author 정보로 들어가는
  - user.name, user.email 을 보통 글로벌로 설정
  - Open Git Log 버튼을 누르면 아래 설명을 터미널창에서 실행
  - --global : 사용자 컴퓨터 전역적인 설정

```
git config --global user.email "you@example.com"

git config --global user.name "your name"
```

# 원격 저장소에서 다운로드

## git pull

- 원격 저장소의 commit log 와 로컬 저장소의 commit log 가 다를때
  - git push error 발생
    - git pull 로 commit log 동기화 시킨 후 다시 파일 수정해서 push

- git fetch + git merger
  - 원격 저장소 commit 으로 동기화 하고 commit 을 merge

## git fetch

- 원격 저장소의 branch 와 commit 들을 로컬 저장소와 동기화
  - merge 는 하지 않음

## git clone

- 초기 프로젝트 시ㅣ 수행하는 명령어
  - 원격 저장소와 정확하게 일치하는 로컬 저장소가 만들어짐
    - 프로젝트 진행 중 clone 은 위험한 작업 : 모든 작업 내용이 없어질 수도 있음
